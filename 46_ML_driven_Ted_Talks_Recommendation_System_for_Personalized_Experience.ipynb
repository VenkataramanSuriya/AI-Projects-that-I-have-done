# **46. ML-driven Ted Talks Recommendation System for Personalized Experience**

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv('/content/ted_main.csv')
```

**Combine relevant text features into a single column**

```python
ted_data = data
```

```python
text_columns = ['description', 'tags', 'speaker_occupation']
ted_data['combined_text'] = ted_data[text_columns].fillna('').agg(' '.join, axis=1)
```

**Initialize and fit TF-IDF vectorizer**

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(ted_data['combined_text'])
```

**Compute cosine similarity matrix**

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

**Function to recommend talks based on talk ID**

```python
def recommend_talks(talk_id, num_recommendations=5):
    sim_scores = list(enumerate(cosine_sim[talk_id]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_similar_talks = sim_scores[1:num_recommendations+1]
    talk_indices = [x[0] for x in top_similar_talks]
    return ted_data.iloc[talk_indices]
```

**Test the recommendation system for a specific talk ID (e.g., talk at index 0)**

```python
recommended_talks = recommend_talks(0, num_recommendations=5)
print(recommended_talks[['title', 'main_speaker', 'views']])
```
