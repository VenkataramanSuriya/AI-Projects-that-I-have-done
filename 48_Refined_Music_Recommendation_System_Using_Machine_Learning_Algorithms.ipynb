```python
from flask import Flask, request, jsonify
from surprise import Dataset, Reader, SVD
import pandas as pd
```

```python
data = pd.read_csv('/content/Music Recomendation System.csv')
```

```python
data
```

**Handling missing values**

```python
data['tempo'].fillna(data['tempo'].mean(), inplace=True)
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')
```

```python
data['artists'] = data['artists'].str.replace('[', '').str.replace(']', '').str.replace("'", '').str.split(', ')
```

```python
print(data.head())
```

**Train your model**

```python
reader = Reader(rating_scale=(0, 1))
data_surprise = Dataset.load_from_df(data[['user', 'item', 'rating']], reader)
trainset = data_surprise.build_full_trainset()
model = SVD()
model.fit(trainset)
```

**API Endpoint for Recommendations**

```python
@app.route('/recommend', methods=['GET'])
def recommend():
    user_id = request.args.get('user_id')
    num_recommendations = int(request.args.get('num_recommendations', 5))
```

**Get recommendations for the user**

```python
recommendations = model.recommend(user_id, num_recommendations)
```

**Format recommendations and return as JSON**

```python
recommended_items = [r[0] for r in recommendations]
    return jsonify({'user_id': user_id, 'recommendations': recommended_items})
```
